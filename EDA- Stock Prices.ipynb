```python
# Importing the required packages
import yfinance as yf
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt

#defining end date for stock prices as current date
current_date = dt.datetime.now()

# Get the data for the TESLA
data = yf.download('TSLA','2010-01-01',current_date)

# Plot the close price of TESLA
plt.figure(figsize=(10, 7))
plt.title("TESLA- Chart View", fontsize=14)
data['Adj Close'].plot()
plt.show()
```

```python

# Define the ticker list
import pandas as pd
tickers_list = ['AAPL', 'META', 'NVDA', 'MSFT', 'GOOG', 'NFLX','TSLA']

# Fetch the data
import yfinance as yf
data = yf.download(tickers_list,'2010-1-1')['Adj Close']

# Print first 10 rows of the data
print(data.head(10))
```

```python

# Plot the close prices of the Stocks
((data.pct_change()+1).cumprod()).plot(figsize=(10, 7))

# Show the legend
plt.legend()

# Define the label for the title of the figure
plt.title("Returns", fontsize=18)

# Define the labels for x-axis and y-axis
plt.ylabel('Stock Cumulative Returns', fontsize=16)
plt.xlabel('Year', fontsize=16)

# Plot the grid lines
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.3)
plt.show()
```
